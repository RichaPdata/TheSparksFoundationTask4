```python
#Importing the Dataset


data = datasets.load_iris()
```

```python
data1 = pd.DataFrame(data.data, columns=data.feature_names)
```

Split Dataset into Train & Test Sets

Important points: 
1. 'data1' and 'data2' are the datasets we want to split.
2. 'test_size=0.33' indicates that 33% of the data will be used as the test set, and the remaining 67% will be the training set.
3. 'random_state=42' is used to ensure reproducibility. Setting a random seed (in this case, 42) ensures that if you run the code multiple times, you will get the same split.

After running this code, we will have four sets of data:

1. data1_train: Training set for data1
2. data1_test: Testing set for data1
3. data2_train: Training set for data2
4. data2_test: Testing set for data2

```python
from sklearn.model_selection import train_test_split
```

```python
data1_train, data1_test, data2_train, data2_test = train_test_split(data1, data2, test_size=0.33, random_state=42)
```

Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data_dtc = DecisionTreeClassifier()
data_dtc.fit(data1_train,data2_train)

print('Decision Tree Classifer Created Successfully')
```

The predict method is then used to make predictions on the testing set (data1_test), and the predicted values are stored in data2_predict. 
These predicted values are the model's output for the given input features.

```python
data2_predict = data_dtc.predict(data1_test)
data2_predict
```

Constructing confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

A confusion matrix is a table that is often used to describe the performance of a classification model.
The confusion_matrix function returns a 2D array where the ROWS represent the 'Actual Class' and the COLUMNS represent the Predicted Class. 
The elements of the matrix are the counts of instances.

Now lets try to generate a confusion matrix using the predicted values (data2_predict) and the actual values (data2_test).

```python
confusion_matrix(data2_test, data2_predict)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

Visualize the Decision tree

```python
Now we will use the 'plot_tree' function from scikit-learn to visualize the decision tree. 
This is a useful way to understand the structure and decisions made by the trained decision tree classifier.
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(data_dtc, feature_names = fn, class_names = cn, filled = True);
```
